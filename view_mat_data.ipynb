```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# .matãƒ•ã‚¡ã‚¤ãƒ«ã®ãƒ‘ã‚¹ã‚’æŒ‡å®š
file_path = 'output/MSE_heatmap_data(N=200, T=300, mu=0.1, ens=100, xi=0).mat'

# ãƒ•ã‚¡ã‚¤ãƒ«ã‚’èª­ã¿è¾¼ã‚€
mat_data = scipy.io.loadmat(file_path)

# å¤‰æ•°åã¨å½¢çŠ¶ã‚’ç¢ºèª
print("ğŸ“Š .matãƒ•ã‚¡ã‚¤ãƒ«ã®å†…å®¹:")
print("=" * 60)
for key in mat_data.keys():
    if not key.startswith('__'):  # ãƒ¡ã‚¿ãƒ‡ãƒ¼ã‚¿ã‚’é™¤å¤–
        value = mat_data[key]
        print(f"\nå¤‰æ•°å: {key}")
        print(f"  å‹: {type(value)}")
        if hasattr(value, 'shape'):
            print(f"  å½¢çŠ¶: {value.shape}")
            print(f"  ãƒ‡ãƒ¼ã‚¿å‹: {value.dtype}")
```

## ãƒ‡ãƒ¼ã‚¿ã®è©³ç´°è¡¨ç¤ºã¨çµ±è¨ˆæƒ…å ±

```python
# ã™ã¹ã¦ã®å¤‰æ•°åã‚’å–å¾—ï¼ˆãƒ¡ã‚¿ãƒ‡ãƒ¼ã‚¿é™¤ãï¼‰
data_keys = [key for key in mat_data.keys() if not key.startswith('__')]

if data_keys:
    # æœ€åˆã®ãƒ‡ãƒ¼ã‚¿å¤‰æ•°ã‚’å–å¾—
    main_key = data_keys[0]
    data_array = mat_data[main_key]
    
    print(f"âœ… å¤‰æ•° '{main_key}' ã®è©³ç´°:")
    print(f"å½¢çŠ¶: {data_array.shape}")
    print(f"\nåŸºæœ¬çµ±è¨ˆ:")
    print(f"  æœ€å°å€¤: {np.min(data_array)}")
    print(f"  æœ€å¤§å€¤: {np.max(data_array)}")
    print(f"  å¹³å‡å€¤: {np.mean(data_array)}")
    print(f"  æ¨™æº–åå·®: {np.std(data_array)}")
    
    # ãƒ‡ãƒ¼ã‚¿ã®ä¸€éƒ¨ã‚’è¡¨ç¤º
    print(f"\nãƒ‡ãƒ¼ã‚¿ã®æœ€åˆã®æ•°è¦ç´ :")
    print(data_array.flat[:10])  # æœ€åˆã®10è¦ç´ 
```

## è¡¨å½¢å¼ã§ãƒ‡ãƒ¼ã‚¿ã‚’è¡¨ç¤ºï¼ˆMATLABã‚¹ã‚¿ã‚¤ãƒ«ï¼‰

```python
# S1_values, S2_values, MSE_matrixã‚’å–å¾—
S1_values = mat_data['S1_values'].flatten()
S2_values = mat_data['S2_values'].flatten()
MSE_matrix = mat_data['MSE_matrix']

print(f"ğŸ“‹ ãƒ‡ãƒ¼ã‚¿ã®æ¦‚è¦:")
print(f"S1ã®ç¯„å›²: {S1_values[0]:.2f} ~ {S1_values[-1]:.2f} ({len(S1_values)}ç‚¹)")
print(f"S2ã®ç¯„å›²: {S2_values[0]:.2f} ~ {S2_values[-1]:.2f} ({len(S2_values)}ç‚¹)")
print(f"MSEè¡Œåˆ—ã®ã‚µã‚¤ã‚º: {MSE_matrix.shape}")

# DataFrameã¨ã—ã¦è¡¨ç¤ºï¼ˆæœ€åˆã®10x10ã®ã¿ï¼‰
df = pd.DataFrame(
    MSE_matrix[:10, :10],
    index=[f'S2={s:.2f}' for s in S2_values[:10]],
    columns=[f'S1={s:.2f}' for s in S1_values[:10]]
)

print("\nğŸ“Š MSEè¡Œåˆ—ï¼ˆæœ€åˆã®10x10ï¼‰:")
print(df.to_string(float_format=lambda x: f'{x:.6e}'))
```

## å¯¾è©±çš„ãªãƒ‡ãƒ¼ã‚¿æ¢ç´¢ï¼ˆå®Œå…¨ãªè¡¨è¡¨ç¤ºï¼‰

```python
# å…¨ãƒ‡ãƒ¼ã‚¿ã‚’DataFrameã¨ã—ã¦è¡¨ç¤º
df_full = pd.DataFrame(
    MSE_matrix,
    index=[f'{s:.2f}' for s in S2_values],
    columns=[f'{s:.2f}' for s in S1_values]
)

# è¡Œã¨åˆ—ã«ãƒ©ãƒ™ãƒ«ã‚’è¨­å®š
df_full.index.name = 'S2'
df_full.columns.name = 'S1'

print("ğŸ“Š å®Œå…¨ãªMSEè¡Œåˆ—:")
print(f"ã‚µã‚¤ã‚º: {df_full.shape[0]} è¡Œ Ã— {df_full.shape[1]} åˆ—")
print("\næœ€åˆã®20è¡ŒÃ—20åˆ—ã‚’è¡¨ç¤º:")
display(df_full.iloc[:20, :20])  # Jupyterç’°å¢ƒã§è¡¨å½¢å¼è¡¨ç¤º

# CSVã¨ã—ã¦ä¿å­˜ã™ã‚‹ã“ã¨ã‚‚å¯èƒ½
# df_full.to_csv('output/MSE_matrix_full.csv')
```

```python
# ãƒ’ãƒ¼ãƒˆãƒãƒƒãƒ—ã‚’æç”»
plt.figure(figsize=(12, 10))
im = plt.imshow(MSE_matrix, aspect='auto', origin='lower', 
                extent=[S1_values[0], S1_values[-1], S2_values[0], S2_values[-1]],
                cmap='jet')
plt.colorbar(im, label='MSE')
plt.xlabel('S1', fontsize=14)
plt.ylabel('S2', fontsize=14)
plt.title('MSE Heatmap', fontsize=16)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ç‰¹å®šã®ãƒã‚¤ãƒ³ãƒˆã§ã®MSEå€¤ã‚’ç¢ºèª
print("\nğŸ” ç‰¹å®šã®ãƒã‚¤ãƒ³ãƒˆã§ã®MSEå€¤:")
for i in [0, len(S1_values)//4, len(S1_values)//2, -1]:
    for j in [0, len(S2_values)//4, len(S2_values)//2, -1]:
        print(f"S1={S1_values[i]:.2f}, S2={S2_values[j]:.2f}: MSE={MSE_matrix[j, i]:.6e}")
```
